```python
import matplotlib.pyplot as plt
import numpy as np

# Data extracted from the classification report
labels = ['0', '1']  # Classes
metrics = ['Precision', 'Recall', 'F1-Score']
data = [
    [0.75, 0.85, 0.79],  # Metrics for class 0
    [0.81, 0.69, 0.74]   # Metrics for class 1
]

# Create a bar graph
x = np.arange(len(metrics))  # Positions for metrics
width = 0.35  # Width of bars

fig, ax = plt.subplots()
bars1 = ax.bar(x - width/2, data[0], width, label='Class 0')
bars2 = ax.bar(x + width/2, data[1], width, label='Class 1')

# Add labels, title, and legend
ax.set_xlabel('Metrics')
ax.set_ylabel('Scores')
ax.set_title('Classification Report Metrics')
ax.set_xticks(x)
ax.set_xticklabels(metrics)
ax.legend()

# Display the scores on the bars
for bars in [bars1, bars2]:
    for bar in bars:
        height = bar.get_height()
        ax.annotate(f'{height:.2f}',
                    xy=(bar.get_x() + bar.get_width() / 2, height),
                    xytext=(0, 3),  # Offset for text
                    textcoords="offset points",
                    ha='center', va='bottom')

# Show the graph
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

# Accuracy value from the classification report
accuracy = 0.77

# Create the bar graph
plt.bar(['Accuracy'], [accuracy], color='skyblue')
plt.ylim(0, 1)  # Set the y-axis limit from 0 to 1
plt.title('Model Accuracy')
plt.ylabel('Accuracy')
plt.text(0, accuracy + 0.02, f'{accuracy:.2f}', ha='center')  # Display the value on top of the bar

# Show the graph
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Data from the table
models = ['Logistic Regression', 'Support Vector Machine (SVC)', 'K-Nearest Neighbor']
accuracies = [84.70, 80.44, 77.20]  # Accuracy values

# Set a beautiful style using seaborn
sns.set_style("whitegrid")
plt.figure(figsize=(8, 5))  # Set the figure size

# Create the bar graph
sns.barplot(x=models, y=accuracies, palette="Blues_d")

# Add accuracy values on top of the bars
for i, accuracy in enumerate(accuracies):
    plt.text(i, accuracy + 0.5, f'{accuracy:.2f}%', ha='center', fontsize=12)

# Customize the graph
plt.title('Model Accuracies', fontsize=16, fontweight='bold')
plt.xlabel('Models', fontsize=12)
plt.ylabel('Accuracy (%)', fontsize=12)
plt.ylim(70, 90)  # Set y-axis limits for better visualization
plt.xticks(rotation=10, fontsize=10)
plt.yticks(fontsize=10)

# Display the graph
plt.tight_layout()
plt.show()
```
